```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from random import randint
def bsearch_game(val):
    tries = 0
    found = False
    guess = randint(0, val)
    if guess == val:
        found = True
    elif guess == 0:
        found = True
    else:
        low, high = 0,  val
        mid = ((high-low)//2)
        while not found:
            tries += 1
            if guess == mid:
                found = True
            elif guess > mid:
                low = mid
                mid = ((high-low)//2) + mid
            else:
                high = mid
                mid = ((high-low)//2) + low
    return tries

values, tries = [], []
for i in range(0,100000,10):
    values.append(i)
    tries.append(bsearch_game(i))
    
```
